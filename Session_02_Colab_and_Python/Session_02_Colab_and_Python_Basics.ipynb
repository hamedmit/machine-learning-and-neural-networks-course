4. NumPy

````python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("جمع:", a + b)
print("میانگین a:", np.mean(a))
````

5. Pandas

````python
import pandas as pd

data = {
    "نام": ["علی", "مریم", "رضا"],
    "نمره": [18, 15, 19]
}

df = pd.DataFrame(data)
df
print("میانگین نمرات:", df["نمره"].mean())
````

6. Matplotlib

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, marker='o')
plt.title("نمودار ساده")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
````

7. تمرین عملی — تحلیل نمرات

````python
import pandas as pd

students = {
    "نام": ["علی", "نگار", "سینا", "مریم"],
    "نمره": [14, 19, 12, 17]
}

df = pd.DataFrame(students)
df
print("بیشترین نمره:", df["نمره"].max())
print("میانگین:", df["نمره"].mean())
````
