```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Ø¯Ø§Ø¯Ù‡ Ø¢Ù…ÙˆØ²Ø´ÛŒ
X = np.array([[60], [80], [120]])
y = np.array([300, 400, 600])

model = LinearRegression()
model.fit(X, y)

pred = model.predict([[100]])
print("Ù‚ÛŒÙ…Øª Ù¾ÛŒØ´â€ŒØ¨ÛŒÙ†ÛŒâ€ŒØ´Ø¯Ù‡:", pred[0], "Ù…ÛŒÙ„ÛŒÙˆÙ† ØªÙˆÙ…Ø§Ù†")
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, label='Ø¯Ø§Ø¯Ù‡ ÙˆØ§Ù‚Ø¹ÛŒ')
plt.plot(X, model.predict(X), color='red', label='Ù…Ø¯Ù„')

plt.xlabel("Ù…Ø³Ø§Ø­Øª Ø®Ø§Ù†Ù‡ (Ù…ØªØ±)")
plt.ylabel("Ù‚ÛŒÙ…Øª (Ù…ÛŒÙ„ÛŒÙˆÙ† ØªÙˆÙ…Ø§Ù†)")
plt.legend()
plt.show()
```
